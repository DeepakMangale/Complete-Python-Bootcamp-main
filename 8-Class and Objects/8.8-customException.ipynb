Custom exception (Raise and Throw an exception)

```python
class Error(Exception):
    pass


class dobException(Error):
    pass 
```

```python
year=int(input("enter the dob"))
age=2025-year

try:
    if age <= 30 and age >= 20:
        print("valid age can apply for exam ")
    else:
        raise dobException
except dobException:
    print("sry,age is not greater than 20 or less than 30")
```
