####  Classes and Objects
Object-Oriented Programming (OOP) is a programming paradigm that uses "objects" to design applications and computer programs. OOP allows for modeling real-world scenarios using classes and objects. This lesson covers the basics of creating classes and objects, including instance variables and methods.

```python
## a class is a blue print for creating objects. attributes , methods


## class and class name = Car ,, blueprint of creating a object 

class Car:
    pass           ## empty Car 


audi=Car()  # obj,type of car
bmw=Car()

print(type(audi))
```

```python
print(audi)
print(bmw)
```

```python
audi.windows=4


print(audi.windows)
```

```python
## instance variable and methods 
class Dog:
## constructor __init__ 
    def __init__(self,name,age):
        self.name=name
        self.age=age

## create objects 
dog1=Dog("buddy",3)
print(dog1)
print(dog1.name)
print(dog1.age)
```

```python
dog2=Dog("lucy",4)
print(dog2)
print(dog2.name)
print(dog2.age)
```

```python
## define a class with instance methods

class Dog:
    def __init__(self,name,age):
        self.name=name
        self.age=age
    
    def bark(self):
        print(f"{self.name} says woof ")
```

```python
dog1=Dog("buddy",3)
dog1.bark()
```

```python
dog2=Dog("lucy",6)
dog2.bark()
```

```python
## modeling a bank account 

## define a class for bank account 
class BankAccount:
    def __init__(self,owner,balance=0):
        self.owner=owner
        self.balance=balance
    def deposit(self,amount):
        self.balance+=amount
        print(f"{amount} is deposit . new balance is {self.balance}")

    def withdraw(self,amount):
        if amount>self.balance:
            print("low money")
        else:
            self.balance-=amount
            print(f"{amount} is withdraw . new balance {self.balance}")
    def get_balance(self):
        return self.balance
    
    # create an account (object)

account=BankAccount("sumit",5000)
print(account.balance)
```

```python
## call isntance method 
account.deposit(100)
```

```python
account.withdraw(300)
```

```python
print(account.get_balance())
```
