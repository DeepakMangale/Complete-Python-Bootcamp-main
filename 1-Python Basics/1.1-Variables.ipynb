### Variables
Variables are fundamental elements in programming used to store data that can be referenced and manipulated in a program. In Python, variables are created when you assign a value to them, and they do not need explicit declaration to reserve memory space. The declaration happens automatically when you assign a value to a variable.

```python
## Declaring and Assigning Variables

Age=32
Height=6.1
Name="sumit"
is_Student=True

## printing the variables

print(f"Name:{Name}")
print(f"Height:{Height}")
print(f"age:{Age}")
print(is_Student)
```

```python
## Naming Conventions
## Variables names should be descriptive 
## they must start with letter or '_' and contain letter,number and underscore
## variables names are case sensitive

#valid variables name 
first_name="sumit"
last_name="mangale"
```

```python
#Understanding variable types

Age=32
Height=6.1
Name="sumit"
is_Student=True

## type checking 

print(type(Age))
print(type(Height))
print(type(Name))
print(type(is_Student))
```

```python
## conversion of type 
age = 23 
print(type(age))

## convert the type from int to string

age_str=str(age)
print(age_str)
print(type(age_str)) 
```

```python
num_flt=23.12
print(type(num_flt))

num_str=int(num_flt)
print(type(num_str))
```

```python
## Dynamic type 
## python allows the type of a variables to change as the program executes

var=10  #int
print(type(var))

var="sumit"  #str
print(type(var))

var=3.14  #float
print(type(var))

```
