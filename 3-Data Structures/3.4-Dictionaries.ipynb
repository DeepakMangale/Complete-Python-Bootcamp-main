##### Introduction to Dictionaries

Dictionaries are unordered collections of items. They store data in key-value pairs.
Keys must be unique and immutable (e.g., strings, numbers, or tuples), while values can be of any type.

```python
## Creating Dictionaries
empty_dict={}
print(type(empty_dict))
```

```python
## creating dictonaries

empty_dict={}
print(type(empty_dict))
```

```python
student={"name":"sumit","age":21,"grade":24}
print(student)
print(type(student))
```

```python
## accessing dict elements

student={"name":"sumit","age":21,"grade":"A"}
print(student)
```

```python
# student["grade"]  
print(student["grade"])
print(student["age"])
```

```python
## accessing using get() method 

print(student.get('grade'))
print(student.get('number'))
```

```python
## modifying dictonary elements 
## dictionary are mutable ,so we can add, update or delete elements 

print(student)
```

```python
student["age"]=22
print(student)
student["address"]="kolhapur"
print(student)
```

```python
## dictionary methods 

keys=student.keys()
print(keys)
value=student.values()
print(value)

item=student.items()
print(item)
```

```python
## shallow copy

student_copy=student
print(student)
print(student_copy)
```

```python
student["name"]="sumit_copy"
print(student)
print(student_copy)
```

```python
student_copy1=student.copy()
print(student_copy1)
print(student)
```

```python
student["name"]="sumit3"
print(student)
print(student_copy1)
```

```python
## iterating over dictionary 
## you can use loop to iterate over dict,keys,values,or items

## iterating over keys 
for keys in student.keys():
    print(keys)
```

```python
## iterating over keys 
for values in student.values():
    print(values)
```

```python
## iterate over key value pairs 

for key,value in student.items():
    print(f"{key}:{value}")
```

```python
## nested dict 

students={
    "student1":{"name":"Krish","age":32},
    "student2":{"name":"sumit","age":42}
}
print(students)
```

```python
## access nested dict elements 
print(students["student2"]["name"])
print(students["student2"]["age"])
```

```python
## iterating over nested dictionaries 
for st_id,st_info in students.items():
    print(f"{st_id}:{st_info}")
    for key,value in st_info.items():
        print(f"{key}:{value}")
```

```python
## dictionary comphrehension 

square={x:x**2 for x in range(10)}
print(square)
```

```python
## conditional dict compherhension 

evens={x:x**2 for x in range(15) if x%2==0}
print(evens)
```

```python
## practice example 

# use a dict to count a frequency of element in list 

numbers=[1,2,3,4,4,5,5,5,6,7,8,0,1,9,0]
frequency={}

for number in  numbers:
    if number in frequency:
        frequency[number]+=1
    else:
        frequency[number]=1
print(frequency)
```

```python
## merge 2 dict into 1 

dict1={"a":1,"b":2}
dict2={"b":3,"c":4}
merge_dict={**dict1,**dict2}
print(merge_dict)
```
