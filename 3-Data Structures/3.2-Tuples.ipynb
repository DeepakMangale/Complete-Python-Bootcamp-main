##### Introduction to Tuples
Explanation:

Tuples are ordered collections of items that are immutable.
They are similar to lists, but their immutability makes them different.


```python
## creating a tuple 


empty_tuple=()
print(empty_tuple)
print(type(empty_tuple))
```

```python
list=list()
print(type(list))
tpl=tuple()
print(type(tpl))
```

```python
## packing of tuple 

pack_tpl=1,"hello",3.14
print(pack_tpl)
```

```python
## unpacking of tuple 

a,b,c=pack_tpl
print(a)
print(b)
print(c)
```

```python
## unpacking with *

numbers=(1,2,3,4,5,6)
first,*middle,last=numbers
print(first)
print(middle)
print(last)
```

```python
## nested tuple 
## nested list 

## list inside list 

lst=[[1,2,3,4,],[5,6,7,8],[1,"hello",3.14,"c"]]
# lst[0] ## take [1,2,3,4]
lst[0][2]  ## take the element inside that 3
```

```python

#tuple in list 


lst=[[1,2,3,4,],[5,6,7,8],(1,"hello",3.14,"c")]
lst[2][0:3]
```

```python
## nested tuple 

nes_tuple=((1,2,3),("a","s","d"),(True,False))
## access the element inside a tuple 

print(nes_tuple[0])
print(nes_tuple[1][2])
```

```python
## iterating over nested tuple 
for sub_tpl in nes_tuple:
    for item in sub_tpl:
        print(item,end=" ")
    print()
    
```
