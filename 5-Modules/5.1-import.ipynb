Importing Modules in Python:Modeles and Packages

in python , modules and packages help organize and reuse code . here's a comprehensive guide on how to import them 

```python
import math
math.sqrt(16)
```

```python
from math import sqrt,pi
print(sqrt(16))
print(sqrt(25))
print(pi)
```

```python
import numpy as np
```

```python
np.array([1,2,3,4])
```

```python
from math import * 

print(sqrt(16))
print(pi)
```

```python
from package.maths import addition
addition(2,3)
```

```python
## another way to call 

from package import maths
maths.addition(2,3)
```

```python
from package.maths import substraction
substraction(6,3)
```
