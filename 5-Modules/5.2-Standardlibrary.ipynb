##### Standard Library Overview
Python's Standard Library is a vast collection of modules and packages that come bundled with Python, providing a wide range of functionalities out of the box. Here's an overview of some of the most commonly used modules and packages in the Python Standard Library.

```python
import array
arr=array.array('i',[1,2,3,4])
print(arr)
```

```python
import math
print(math.sqrt(16))
print(math.pi)
```

```python
## random  

import random 
print(random.randint(1,10))
# it will give any one random int number between 1 to 10 

print(random.choice(['apple','banana','cherry']))
# it will choice any random fruit between them 
```

```python
## data serialization 
import json
data={'name':'sumit','age':21}

## json.dumps  will change it from dict.. to str
json_str=json.dumps(data)
print(json_str)
print(type(json_str))

## json.loads   will again change it into dict
past_data=json.loads(json_str)
print(past_data)
print(type(past_data))
```

```python
## csv 
## how to make a csv file  

import csv
with open('example.csv',mode='w',newline='')as file:
    writer=csv.writer(file)
    writer.writerow(['name','age'])
    writer.writerow(['sumit','32'])

with open('example.csv',mode='r')as file:
    reader=csv.reader(file)
    for row in reader:
        print(row)
```

```python
## datetime
from datetime import datetime,timedelta

now=datetime.now()
print(now)

yesterday=now-timedelta(days=1)

print(yesterday)
```

```python
## time
import time
print(time.time())
time.sleep(2)
print(time.time())
```

```python
## Regular expresiion
import re

pattern=r'\d+'
text='There are 123 apples 456'
match=re.search(pattern,text)
print(match.group())
```
