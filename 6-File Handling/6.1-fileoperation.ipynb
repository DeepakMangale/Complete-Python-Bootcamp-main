#### File Operation- Read And Write Files

File handling is a crucial part of any programming language. Python provides built-in functions and methods to read from and write to files, both text and binary. This lesson will cover the basics of file handling, including reading and writing text files and binary files.

```python
## read a whole file 
with open('example.txt','r') as file:
    content=file.read()
    print(content)
```

```python
## read a file line by line 

with open('example.txt','r') as file :
    for line in file:
        print(line.strip())## strip() removes the newline character
```

```python
## open a file to writing a file(overwriting)

with open('example.txt','w') as file:  ## 'w' means write mode 
    file.write('hello world\n')
    file.write('this is new line this is new content  \n')
    file.write('adshjfbhdbfhbdshbvrksdbkvsdhkvfs \n')
    file.write('adshjfbhdbfhbdshbvrksdbkvsdhkvfs \n')

    ## now beacuse of this the old content will gone and this two will display 
     
```

```python
## write a file (without overwriting )
with open ('example.txt','a') as file:  ## 'a' means append mode 
    file.write('append op is taking place \n')
```

```python
# writing a list of line to a file
lines=['first line \n','second line \n','third line']
with open('example.txt','a') as file:
    file.writelines(lines)
    
```

```python
## how to work with binary files 

# writing to a binary file 
data = b'\x00\x01\x02\x03\x04afd'  ## hexza char
with open('example.bin','wb') as file: ## 'wb'-write bite
    file.write(data)
```

```python
## reading a binary file 

with open('example.bin','rb') as file:
    content = file.read()
    print(content)
```

```python
## read the content from source text file and wrtie to a destination text file 
# copying a text file 

with open('example.txt','r') as source_file:
    content = source_file.read()

with open('destination_file','w') as destination_file:
    destination_file.write(content)
```
