```python
## if statement

age = 20
if age >= 18:
    print("u can vote")
else:
    print("not able to vote") 
```

```python
##  elif statement

age = 20
if age <= 12:
    print("child")
elif age <=18:
    print("teenager")

else:
    print("adult") 
```

```python
## nested condtional statement

num = int(input("enter your no."))
if num>0:
    print("num is positive")
    if num%2==0:
        print("num is even")
    else:
        print("num is odd")
else:
    print("num is negative")
```

```python
## check if the year is leap year or not 

year = int(input("enter year: "))
if year%4==0:
    if year%100==0:
        if year%400==0:
                print("this is leap year")
        else:
            print("not a leap year")
    else:
        print("leap year")
else:
    print("not a leap year")
```

```python
## calculate

num1 = float(input("enter 1 number"))
num2 = float(input("enter 1 number"))

op = input("enter op (+,-,*,/):")

if op == '+':
    result = num1 + num2
elif op == '-':
    result = num1 - num2 
elif op == '*':
    result = num1 * num2 
elif op == '/':
    if num2!=0:
        result = num1 / num2
    else:
        result = "error! division by zero"
else:
    print("invalid ")
print(result)
```
