## while loop

```python
## loop will run and output will be given until the condition get false
```

```python
count = 0
while count<8:
    print(count)
    count=count+1
```

```python
## break statement 

for i in range(10):
    if i==6:
        break
    print(i)
```

```python
## continue statement

for i in range(20):
    if i%2==0:
        continue
    print(i)
```

```python
for i in range(3):
    for j in range(2):
        print(f"i:{i} and j:{j}")
```

## examples

```python
## while loop 
## cal the n natural number 

n=10
sum=0
count=1

while count<=n:
    sum=sum+count
    count=count+1
print(sum)
```

```python
## print  prime numbers bet 1 - 100

for num in range(1,101):
    if num>1:
        for i in range(2,num):
            if num%i==0:
                break
        else:
                print(num)
        
```
